Data pre processing

```python
#encoding categorical features
#encoding sex category
insurance_dataset.replace({'sex':{'male':0,'female':1}},inplace=True)
#encoding smoker columm
insurance_dataset.replace({'smoker':{'yes':0,'no':1}},inplace=True)
#encoding region column
insurance_dataset.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
```

```python
#splitting the featires and target
X = insurance_dataset.drop(columns='charges',axis=1)
Y = insurance_dataset['charges']
```

```python
print (X)
```

```python
print (Y)
```

```python
#splitting the data into traing and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print (X.shape,X_train.shape,X_test.shape)
```

Model training






Linear regression

```python
#LOADING linear regressioon model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
##Model Evaluation

#preditopn on traning data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R squared value
r2_train = metrics.r2_score(Y_train,training_data_prediction)
print ('R squared vale:',r2_train)
```

```python
#prediction of testing data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
#R squaredv value
r2_test = metrics.r2_score(Y_test,test_data_prediction)
print ('R squared vale:',r2_test)
```

```python
#build the predictive system
input_data = (31,1,25.74,0,1,0)
# changing input_data to a numpy array
input_data_as_numpy_array = np.array(input_data)

# reshape the array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = lin_reg_model.predict(input_data_reshaped)
print(prediction)

print('The insurance cost is USD ', prediction[0])
```
